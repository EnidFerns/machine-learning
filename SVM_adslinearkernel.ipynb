```python
df = pd.read_csv('/content/drive/MyDrive/Verzeo ML Feb Batch/Ads.csv')
```

```python
x = df.iloc[:,2:4].values#reshape not required as already 2d#2columns
```

```python
y = df.iloc[:,4].values
```

```python
#y can be 1d but x has to be 2d
```

```python
#train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
#feature scaling is required here
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(x)
```

```python
x#converted to standard normal distributuion
#done cuz age is 2 digitno whereas salary is more
```

```python
#now divide
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, )
```

```python
from sklearn.svm import SVC
```

```python
class_svc = SVC(kernel = 'linear')
```

```python
class_svc.fit(x_train,y_train)
```

```python
y_pred = class_svc.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
#confusion matrix
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
accuracy_score(y_test,y_pred)
```
